We need the size of apps in common unit

```python
#define a function to convert all app sizes into MB
def convert(size):
    if size[-1] == 'k':
        return float(size[:-1])*0.001
    elif size[-1] == 'M':
        return float(size[:-1])
    else:
        return np.NaN
```

```python
clean_df['Size'] = clean_df['Size'].apply(convert)
```

```python
clean_df['Size'].describe()
```

```python
clean_df['Size'].isnull().sum()
```

```python
mean_size = clean_df['Size'].mean()
clean_df['Size'].fillna(mean_size, inplace=True)
```

We replaced the missing size values with mean value.

```python
#function to transform string of numbers to integers
def transform_1(s):
    if s == '0':
        return 0
    else:
        return int(s.replace(',','')[:-1])
```

```python
clean_df['Installs'] = clean_df['Installs'].apply(transform_1)
clean_df['Installs'].head(10)
```

Removed '+' from 'Number of Installs' to make it numeric

```python
clean_df.drop(index=[9148], inplace=True)
```

We need to remove dollar sign from the price 

```python
def transform_2(s):
    if s == '0':
        return 0
    else:
        return float(s[1:])
```

```python
clean_df['Price'] = clean_df['Price'].apply(transform_2)
clean_df['Price']
```

```python
Categories = clean_df['Category'].unique()
```

```python
Category_df = pd.DataFrame(index=Categories)
def def_df(cols = ['Rating', 'Size', 'Price', 'Installs']):
    Grouped_df = clean_df.groupby('Category')
    Category_df['Count'] = Grouped_df.size()
    Category_df[[cols[0] +'_mean', cols[1] +'_mean', cols[2] +'_mean', cols[3] +'_mean']] = Grouped_df.mean()[['Rating', 'Size', 'Price', 'Installs']]
    Category_df[[cols[0] +'_std', cols[1] +'_std', cols[2] +'_std', cols[3] +'_std']] = Grouped_df.std()[['Rating', 'Size', 'Price', 'Installs']]
def_df()
```
